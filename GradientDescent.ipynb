```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

class LinearRegressionGD:
    """Gradient Descent for Univariate Linear Regression"""
    
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta0 = 0  # Intercept
        self.theta1 = 0  # Slope
        self.cost_history = []
        self.theta0_history = []
        self.theta1_history = []
        
    def predict(self, X):
        """Make predictions: y = theta0 + theta1 * X"""
        return self.theta0 + self.theta1 * X
    
    def compute_cost(self, X, y):
        """Compute Mean Squared Error (MSE)"""
        m = len(X)
        predictions = self.predict(X)
        error = predictions - y
        cost = (1/(2*m)) * np.sum(error ** 2)
        return cost
    
    def fit(self, X, y, verbose=False):
        """Train the model using gradient descent"""
        m = len(X)
        self.cost_history = []
        self.theta0_history = []
        self.theta1_history = []
        
        # Initialize parameters
        self.theta0 = 0
        self.theta1 = 0
        
        for i in range(self.n_iterations):
            # Make predictions
            predictions = self.predict(X)
            
            # Compute gradients
            error = predictions - y
            d_theta0 = (1/m) * np.sum(error)
            d_theta1 = (1/m) * np.sum(error * X)
            
            # Update parameters
            self.theta0 -= self.learning_rate * d_theta0
            self.theta1 -= self.learning_rate * d_theta1
            
            # Store history for visualization
            self.theta0_history.append(self.theta0)
            self.theta1_history.append(self.theta1)
            cost = self.compute_cost(X, y)
            self.cost_history.append(cost)
            
            # Print progress
            if verbose and i % 100 == 0:
                print(f"Iteration {i}: Cost = {cost:.4f}, "
                      f"theta0 = {self.theta0:.4f}, theta1 = {self.theta1:.4f}")
        
        return self
    
    def get_parameters(self):
        """Return model parameters"""
        return {"theta0": self.theta0, "theta1": self.theta1}


# Generate synthetic data
def generate_data(n_samples=100, noise_level=10):
    """Generate synthetic linear data with noise"""
    X = np.random.rand(n_samples) * 10  # Features between 0 and 10
    true_theta0 = 2  # True intercept
    true_theta1 = 3  # True slope
    noise = np.random.randn(n_samples) * noise_level
    y = true_theta0 + true_theta1 * X + noise
    return X, y, true_theta0, true_theta1


def plot_regression_results(X, y, model, true_params=None, ax=None):
    """Plot the regression line against the data"""
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot data points
    ax.scatter(X, y, alpha=0.6, label='Data points', color='blue')
    
    # Plot regression line
    x_line = np.linspace(X.min(), X.max(), 100)
    y_pred = model.predict(x_line)
    ax.plot(x_line, y_pred, color='red', linewidth=3, 
            label=f'Regression line: y = {model.theta0:.2f} + {model.theta1:.2f}x')
    
    # Plot true line if provided
    if true_params:
        y_true = true_params['theta0'] + true_params['theta1'] * x_line
        ax.plot(x_line, y_true, '--', color='green', linewidth=2, 
                label=f'True line: y = {true_params["theta0"]} + {true_params["theta1"]}x')
    
    ax.set_xlabel('X')
    ax.set_ylabel('y')
    ax.set_title('Linear Regression Fit')
    ax.legend()
    ax.grid(True, alpha=0.3)
    return ax


def plot_cost_history(model, ax=None):
    """Plot the cost function history during training"""
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))
    
    ax.plot(range(len(model.cost_history)), model.cost_history, 
            color='purple', linewidth=2)
    ax.set_xlabel('Iteration')
    ax.set_ylabel('Cost (MSE)')
    ax.set_title('Cost Function History')
    ax.grid(True, alpha=0.3)
    
    # Add annotation for final cost
    final_cost = model.cost_history[-1]
    ax.annotate(f'Final Cost: {final_cost:.4f}', 
                xy=(0.7, 0.9), xycoords='axes fraction',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.5))
    return ax


def plot_parameter_space(model, X, y, ax=None):
    """Visualize the parameter space and gradient descent path"""
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 8))
    
    # Create a grid of parameter values
    theta0_vals = np.linspace(-5, 10, 100)
    theta1_vals = np.linspace(-1, 7, 100)
    theta0_grid, theta1_grid = np.meshgrid(theta0_vals, theta1_vals)
    
    # Compute cost for each parameter combination
    m = len(X)
    cost_grid = np.zeros_like(theta0_grid)
    
    for i in range(theta0_grid.shape[0]):
        for j in range(theta0_grid.shape[1]):
            predictions = theta0_grid[i, j] + theta1_grid[i, j] * X
            error = predictions - y
            cost_grid[i, j] = (1/(2*m)) * np.sum(error ** 2)
    
    # Plot contour
    contour = ax.contour(theta0_grid, theta1_grid, cost_grid, 
                        levels=20, cmap='viridis', alpha=0.6)
    ax.clabel(contour, inline=True, fontsize=8)
    
    # Plot gradient descent path
    ax.plot(model.theta0_history, model.theta1_history, 
            'r-', linewidth=2, label='Gradient Descent Path')
    ax.scatter(model.theta0_history, model.theta1_history, 
              c=range(len(model.theta0_history)), cmap='Reds', s=30)
    
    # Mark start and end points
    ax.scatter(model.theta0_history[0], model.theta1_history[0], 
              color='green', s=200, marker='*', label='Start', zorder=5)
    ax.scatter(model.theta0, model.theta1, 
              color='red', s=200, marker='*', label='Optimal', zorder=5)
    
    ax.set_xlabel(' (Intercept)')
    ax.set_ylabel(' (Slope)')
    ax.set_title('Gradient Descent in Parameter Space')
    ax.legend()
    ax.grid(True, alpha=0.3)
    return ax


def create_animation(X, y, model_history, true_params=None):
    """Create an animation of gradient descent convergence"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    def update(frame):
        # Clear axes
        ax1.clear()
        ax2.clear()
        
        # Get parameters for this frame
        theta0, theta1 = model_history[frame]
        
        # Plot 1: Regression line
        ax1.scatter(X, y, alpha=0.6, color='blue')
        x_line = np.linspace(X.min(), X.max(), 100)
        y_pred = theta0 + theta1 * x_line
        ax1.plot(x_line, y_pred, 'r-', linewidth=3)
        
        if true_params:
            y_true = true_params['theta0'] + true_params['theta1'] * x_line
            ax1.plot(x_line, y_true, 'g--', linewidth=2)
        
        ax1.set_xlim(X.min() - 1, X.max() + 1)
        ax1.set_ylim(y.min() - 10, y.max() + 10)
        ax1.set_xlabel('X')
        ax1.set_ylabel('y')
        ax1.set_title(f'Iteration {frame}: y = {theta0:.2f} + {theta1:.2f}x')
        ax1.grid(True, alpha=0.3)
        
        # Plot 2: Cost history
        ax2.plot(range(frame+1), model.cost_history[:frame+1], 'purple', linewidth=2)
        ax2.set_xlabel('Iteration')
        ax2.set_ylabel('Cost')
        ax2.set_title('Cost Function Convergence')
        ax2.grid(True, alpha=0.3)
        
        fig.suptitle(f'Gradient Descent Progress (Iteration {frame})', fontsize=14)
    
    # Create animation
    anim = FuncAnimation(fig, update, frames=min(100, len(model_history)), 
                        interval=50, repeat=False)
    plt.close()
    return anim


# Main execution
def main():
    """Main function to run the complete example"""
    print("=" * 60)
    print("GRADIENT DESCENT FOR LINEAR REGRESSION - VISUALIZATION")
    print("=" * 60)
    
    # Step 1: Generate data
    print("\n1. Generating synthetic data...")
    X, y, true_theta0, true_theta1 = generate_data(n_samples=50, noise_level=5)
    print(f"   Generated {len(X)} data points")
    print(f"   True parameters: theta0 = {true_theta0}, theta1 = {true_theta1}")
    
    # Step 2: Create and train model
    print("\n2. Training model with Gradient Descent...")
    model = LinearRegressionGD(learning_rate=0.01, n_iterations=1000)
    model.fit(X, y, verbose=True)
    
    # Step 3: Display results
    print("\n3. Results:")
    params = model.get_parameters()
    print(f"   Learned parameters:")
    print(f"   - theta0 (intercept): {params['theta0']:.4f}")
    print(f"   - theta1 (slope): {params['theta1']:.4f}")
    print(f"   Final cost: {model.cost_history[-1]:.4f}")
    
    # Step 4: Create comprehensive visualization
    print("\n4. Creating visualizations...")
    
    fig = plt.figure(figsize=(16, 12))
    
    # Subplot 1: Regression fit
    ax1 = plt.subplot(2, 2, 1)
    plot_regression_results(X, y, model, 
                           true_params={'theta0': true_theta0, 'theta1': true_theta1}, 
                           ax=ax1)
    
    # Subplot 2: Cost function history
    ax2 = plt.subplot(2, 2, 2)
    plot_cost_history(model, ax=ax2)
    
    # Subplot 3: Parameter space
    ax3 = plt.subplot(2, 2, 3)
    plot_parameter_space(model, X, y, ax=ax3)
    
    # Subplot 4: Error distribution
    ax4 = plt.subplot(2, 2, 4)
    predictions = model.predict(X)
    errors = predictions - y
    ax4.hist(errors, bins=20, edgecolor='black', alpha=0.7, color='orange')
    ax4.axvline(x=0, color='red', linestyle='--', linewidth=2)
    ax4.set_xlabel('Prediction Error')
    ax4.set_ylabel('Frequency')
    ax4.set_title('Error Distribution')
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Step 5: Create and display animation (optional - for Jupyter notebooks)
    try:
        print("\n5. Creating gradient descent animation...")
        # Prepare history for animation
        model_history = list(zip(model.theta0_history, model.theta1_history))
        
        # Create a simpler model for animation with fewer iterations
        model_anim = LinearRegressionGD(learning_rate=0.01, n_iterations=100)
        model_anim.fit(X, y, verbose=False)
        model_anim_history = list(zip(model_anim.theta0_history, model_anim.theta1_history))
        
        anim = create_animation(X, y, model_anim_history, 
                               true_params={'theta0': true_theta0, 'theta1': true_theta1})
        
        # For Jupyter Notebook
        from IPython.display import HTML
        HTML(anim.to_html5_video())
        
        print("   Animation created! Check your notebook for the animation.")
    except Exception as e:
        print(f"   Animation could not be displayed: {e}")
        print("   (This is normal if you're not in a Jupyter notebook)")
    
    # Step 6: Interactive learning rate comparison
    print("\n6. Comparing different learning rates...")
    
    learning_rates = [0.001, 0.01, 0.1, 0.5]
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    axes = axes.flatten()
    
    for idx, lr in enumerate(learning_rates):
        model_lr = LinearRegressionGD(learning_rate=lr, n_iterations=1000)
        model_lr.fit(X, y, verbose=False)
        
        axes[idx].plot(model_lr.cost_history, linewidth=2)
        axes[idx].set_xlabel('Iteration')
        axes[idx].set_ylabel('Cost')
        axes[idx].set_title(f'Learning Rate = {lr}')
        axes[idx].grid(True, alpha=0.3)
        axes[idx].set_yscale('log')  # Log scale to see differences better
        
        # Mark convergence point
        final_cost = model_lr.cost_history[-1]
        axes[idx].annotate(f'Final: {final_cost:.4f}', 
                          xy=(0.6, 0.8), xycoords='axes fraction',
                          bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.5))
    
    plt.suptitle('Effect of Learning Rate on Gradient Descent Convergence', fontsize=14)
    plt.tight_layout()
    plt.show()
    
    print("\n" + "=" * 60)
    print("ANALYSIS:")
    print("=" * 60)
    print(f"1. Optimal parameters found:  = {params['theta0']:.4f},  = {params['theta1']:.4f}")
    print(f"2. True parameters were:  = {true_theta0},  = {true_theta1}")
    print(f"3. Cost reduced from {model.cost_history[0]:.4f} to {model.cost_history[-1]:.4f}")
    print(f"4. Learning rate {model.learning_rate} gave stable convergence")
    print("=" * 60)

# Run the main function
if __name__ == "__main__":
    main()
```
