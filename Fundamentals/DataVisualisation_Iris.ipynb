## Step 1: Load the Iris Dataset

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")
print(iris.head())
```

## Step 2: Correlation Matrix

```python
# Compute correlation matrix
corr_matrix = iris.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Iris Dataset")
plt.show()
```

## Step 3: Relationship Graphs

```python
# Pairplot to visualize relationships
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```

## Step 4: Standard Deviation Visualization

```python
# Compute standard deviations
std_values = iris.std(numeric_only=True)

# Bar plot of standard deviations
plt.figure(figsize=(8,6))
std_values.plot(kind="bar", color="skyblue")
plt.title("Standard Deviations of Iris Features")
plt.ylabel("Standard Deviation")
plt.show()
```

## Step 5: Boxplots for Spread & Variation

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=iris.drop(columns="species"))
plt.title("Boxplots of Iris Features")
plt.show()
```

## Step 6: Compute Standard Deviations by Species

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")

# Group by species and compute standard deviation
species_std = iris.groupby("species").std(numeric_only=True)

print(species_std)
```

```python
## Step 7: Plot Standard Deviations by Species
```

```python
# Bar plot of standard deviations for each species
species_std.plot(kind="bar", figsize=(10,6))
plt.title("Standard Deviations of Iris Features by Species")
plt.ylabel("Standard Deviation")
plt.xlabel("Species")
plt.legend(title="Features")
plt.show()
```

```python
## Step 8: Boxplots for Species Comparison
```

```python
# Boxplots separated by species
plt.figure(figsize=(12,8))
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Sepal Length Distribution by Species")
plt.show()

plt.figure(figsize=(12,8))
sns.boxplot(x="species", y="petal_length", data=iris)
plt.title("Petal Length Distribution by Species")
plt.show()
```

```python
## Step 9: Violin Plots (Optional, for richer visualization)
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(x="species", y="sepal_width", data=iris, inner="quartile")
plt.title("Sepal Width Variability by Species")
plt.show()
```

```python
## Step 1: Gaussian Distribution for Each Feature
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Features to visualize
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

# Plot Gaussian distribution for each feature
for feature in features:
    plt.figure(figsize=(8,6))
    sns.histplot(iris[feature], kde=True, stat="density", bins=20, color="skyblue")
    plt.title(f"Gaussian Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Density")
    plt.show()
```

```python
## Step 2: Gaussian Distribution by Species
```

```python
for feature in features:
    plt.figure(figsize=(8,6))
    sns.histplot(data=iris, x=feature, hue="species", kde=True, stat="density", bins=20)
    plt.title(f"Gaussian Distribution of {feature} by Species")
    plt.xlabel(feature)
    plt.ylabel("Density")
    plt.show()
```

```python
## Step 3: Combined Gaussian Visualization (Optional Dashboard)
```

```python
sns.displot(iris, x="sepal_length", hue="species", kind="kde", fill=True, height=6, aspect=1.5)
plt.title("Gaussian KDE of Sepal Length by Species")
plt.show()

sns.displot(iris, x="petal_length", hue="species", kind="kde", fill=True, height=6, aspect=1.5)
plt.title("Gaussian KDE of Petal Length by Species")
plt.show()
```
