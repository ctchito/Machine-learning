A box plot is a powerful visualization for spotting outliers and understanding the spread of your data. Let‚Äôs break it down and then show you how to implement it in Python.

üìä What is a Box Plot?
‚Ä¢ Box ‚Üí Shows the interquartile range (IQR: Q1 to Q3).
‚Ä¢ Line inside box ‚Üí Median (Q2).
‚Ä¢ Whiskers ‚Üí Extend to data points within 1.5 √ó IQR from Q1 and Q3.
‚Ä¢ Outliers ‚Üí Points beyond the whiskers, plotted individually.
üëâ Outliers are values that lie outside the range:
\text{Lower Bound} = Q1 - 1.5 \cdot IQR, \quad \text{Upper Bound} = Q3 + 1.5 \cdot IQR


## üè† Example: Housing Dataset (Synthetic)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic housing prices
np.random.seed(42)
prices = np.random.normal(200000, 50000, 500)

# Add some extreme outliers
prices = np.append(prices, [600000, 700000, 800000])

# Create DataFrame
df = pd.DataFrame({"price": prices})

# Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x=df["price"])
plt.title("Box Plot of Housing Prices")
plt.show()
```

## Detect Outliers Programmatically

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["price"] < lower_bound) | (df["price"] > upper_bound)]
print("Outliers:\n", outliers)
```
